```python
data["neighbourhood_group"] = data["neighbourhood_group"].astype('category')
data["neighbourhood_group"] = data["neighbourhood_group"].cat.codes

data["neighbourhood"] = data["neighbourhood"].astype('category')
data["neighbourhood"] = data["neighbourhood"].cat.codes

data["room_type"] = data["room_type"].astype('category')
data["room_type"] = data["room_type"].cat.codes



```

```python
data.head()
```

```python
# selecting columns
# data.columns

data = data.drop(['name', 'host_id', 'calculated_host_listings_count'], axis=1)
data.head()
```

```python
# linear regression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler



errors = []
for loop in range(10):
    data = np.array(data)
    np.random.shuffle(data)

    n_train = int(.8 * len(data))
    # n_test = len(data) - n_train

    X_train, Y_train = data[:n_train, :-1], data[:n_train, -1]
    X_test, Y_test = data[n_train:, :-1], data[n_train:, -1] 
    # Create linear regression object

    scaler = StandardScaler()
    scaler.fit_transform(X_train)
    
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    

    regr = linear_model.LinearRegression()
    regr.fit(X_train, Y_train)


    # Make predictions using the testing set
    y_pred_train = regr.predict(X_train)

    print('Mean squared error over train data: %.2f'
          % mean_absolute_error(Y_train, y_pred_train))


    # Make predictions using the testing set
    y_pred = regr.predict(X_test)

    print('Mean squared error over test data: %.2f'
            % mean_absolute_error(Y_test, y_pred))
    #       % mean_squared_error(Y_test, y_pred))
    
    
    errors.append(mean_absolute_error(Y_test, y_pred))

print(f"avg. error: {np.mean(errors)}")    
```

```python
# svm
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler


errors = []
for loop in range(1):
    data = np.array(data)
    np.random.shuffle(data)

    n_train = int(.8 * len(data))
    # n_test = len(data) - n_train

    X_train, Y_train = data[:n_train, :-1], data[:n_train, -1]
    X_test, Y_test = data[n_train:, :-1], data[n_train:, -1] 
    # Create linear regression object

    scaler = StandardScaler()
    scaler.fit_transform(X_train)
    
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    

    svr = SVR(kernel='rbf', gamma=0.1, C=1e1, epsilon=0.2)
    svr.fit(X_train, Y_train)


    # Make predictions using the testing set
    y_pred_train = svr.predict(X_train)

    print('Mean squared error over train data: %.2f'
          % mean_absolute_error(Y_train, y_pred_train))


    # Make predictions using the testing set
    y_pred = svr.predict(X_test)

    print('Mean squared error over test data: %.2f'
            % mean_absolute_error(Y_test, y_pred))
    #       % mean_squared_error(Y_test, y_pred))
    
    
    errors.append(mean_absolute_error(Y_test, y_pred))

print(f"avg. error: {np.mean(errors)}")    
```

```python

# Knn regression
from sklearn import neighbors
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler


errors = []
for loop in range(4):
    data = np.array(data)
    np.random.shuffle(data)

    n_train = int(.9 * len(data))
    # n_test = len(data) - n_train

    X_train, Y_train = data[:n_train, :-1], data[:n_train, -1]
    X_test, Y_test = data[n_train:, :-1], data[n_train:, -1] 
    # Create linear regression object

    scaler = StandardScaler()
    scaler.fit_transform(X_train)
    
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)    

    knn = neighbors.KNeighborsRegressor(n_neighbors=50, weights='distance')

    # Make predictions using the testing set
    y_pred_train = knn.fit(X_train, Y_train).predict(X_train)

    print('Mean squared error over train data: %.2f'
          % mean_absolute_error(Y_train, y_pred_train))


    # Make predictions using the testing set
    y_pred = knn.fit(X_train, Y_train).predict(X_test)

    print('Mean squared error over test data: %.2f'
            % mean_absolute_error(Y_test, y_pred))
    #       % mean_squared_error(Y_test, y_pred))
    
    
    errors.append(mean_absolute_error(Y_test, y_pred))

print(f"avg. error: {np.mean(errors)}")
```

```python

# ridge regression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler


errors = []
for loop in range(4):
    data = np.array(data)
    np.random.shuffle(data)

    n_train = int(.8 * len(data))
    # n_test = len(data) - n_train

    X_train, Y_train = data[:n_train, :-1], data[:n_train, -1]
    X_test, Y_test = data[n_train:, :-1], data[n_train:, -1] 
    # Create linear regression object

    scaler = StandardScaler()
    scaler.fit_transform(X_train)
    
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)


#     regr = linear_model.Ridge(alpha=.2, fit_intercept=True)
    regr = make_pipeline(PolynomialFeatures(2), Ridge(alpha=.3))
    regr.fit(X_train, Y_train)


    # Make predictions using the testing set
    y_pred_train = regr.predict(X_train)

    print('Mean squared error over train data: %.2f'
          % mean_absolute_error(Y_train, y_pred_train))


    # Make predictions using the testing set
    y_pred = regr.predict(X_test)

    print('Mean squared error over test data: %.2f'
            % mean_absolute_error(Y_test, y_pred))
    #       % mean_squared_error(Y_test, y_pred))
    
    
    errors.append(mean_absolute_error(Y_test, y_pred))

print(f"avg. error: {np.mean(errors)}")    
```
